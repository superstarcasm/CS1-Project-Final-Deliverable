```python
print("Welcome!")
def CalcGWA():
  x = 1.50
  GWAnow = float(input("Input your GWA this quarter: "))
  if GWAnow > 5.00 or GWAnow < 1:
    print("Not a valid GWA. Please enter a correct one.")
  else:
    if GWAnow <= 2.5:
      GWANeeded = (x - (GWAnow / 3)) * 1.5
      if GWANeeded == 1:
        print(f"You need {GWANeeded} to get DL.")
      else:
        print(f"You need {GWANeeded} or better to get DL")
    else:
      print("You cannot get DL next quarter.")

CalcGWA()
print("Thank you for trying out the GWhat You Need!")
```
